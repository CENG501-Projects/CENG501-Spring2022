```python
# import libraries here
import matplotlib.pyplot as plt
import json
import numpy as np
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)
```

# Plotting functions

```python
## plot for SGM with different learning schedulers (CIFAR-10, Fashion-MNIST)
def plot_loss_nn(path, file_names, labels, colors, marks, title, plot_num):
    plt.subplot(1,4, plot_num)
    x = [i for i in range(10,210,10)]
    plt.xticks(size = 18)
    plt.yticks(size = 18)
    plt.xlabel('Number of effective passes', fontsize = 18)
    plt.ylabel('Train Loss', fontsize = 18)
    plt.yscale('log')
    plt.title(title)

    for index, file_name in enumerate(file_names):
        full_path = path + file_name + '.json'
        with open(full_path, 'r') as json_file:
            data = json.load(json_file)
            loss_history = data['loss_history']
            y = [np.sum(loss_history[k:k+10])/10 for k in range(0,200,10)]
            plt.plot(x, y, marks[index], color=colors[index], label=labels[index])
    
    plt.legend(loc='best', fontsize=18)
    if plot_num == 4:
        plt.tight_layout()
        plt.show()
    

def plot_loss_non_convex(path, file_names, labels, colors, marks, title, plot_num):
    plt.subplot(1,4, plot_num)
    x = [i for i in range(3,50,3)]
    plt.xticks(size = 18)
    plt.yticks(size = 18)
    plt.xlabel('Number of effective passes', fontsize = 18)
    plt.ylabel('Train Loss', fontsize = 18)
    plt.yscale('log')
    plt.title(title)

    for index, file_name in enumerate(file_names):
        full_path = path + file_name + '.json'
        with open(full_path, 'r') as json_file:
            data = json.load(json_file)
            loss_history = data['loss_history']
            y = [np.sum(loss_history[k:k+3])/3 for k in range(0,47,3)]
            plt.plot(x, y, marks[index], color=colors[index], label=labels[index])
    
    plt.legend(loc='best', fontsize=18)
    if plot_num == 4:
        plt.tight_layout()
        plt.show()
```

```python
## plot for SGM with different learning schedulers (CIFAR-10)
def plot_grad_squares_nn(path, file_names, labels, colors, marks, title, plot_num):
    plt.subplot(1,4, plot_num)
    x = [i for i in range(10,210,10)]
    plt.xticks(size = 18)
    plt.yticks(size = 18)
    plt.xlabel('Number of effective passes', fontsize = 18)
    plt.ylabel('||∇F(w)||²', fontsize = 18)
    plt.yscale('log')
    plt.title(title)

    for index, file_name in enumerate(file_names):
        full_path = path + file_name + '.json'
        with open(full_path, 'r') as json_file:
            data = json.load(json_file)
            grad_history = data['grad_history']
            y = [np.sum(grad_history[k:k+10])/10 for k in range(0,200,10)]
            plt.plot(x, y, marks[index], color=colors[index], label=labels[index])
    
    plt.legend(loc='best', fontsize=18)
    if plot_num == 4:
        plt.tight_layout()
        plt.show()


def plot_grad_squares_non_convex(path, file_names, labels, colors, marks, title, plot_num):
    plt.subplot(1,4, plot_num)
    x = [i for i in range(3,50,3)]
    plt.xticks(size = 18)
    plt.yticks(size = 18)
    plt.xlabel('Number of effective passes', fontsize = 18)
    plt.ylabel('||∇F(w)||²', fontsize = 18)
    plt.yscale('log')
    plt.title(title)
    
    for index, file_name in enumerate(file_names):
        full_path = path + file_name + '.json'
        with open(full_path, 'r') as json_file:
            data = json.load(json_file)
            grad_history = data['grad_history']
            y = [np.sum(grad_history[k:k+3])/3 for k in range(0,47,3)]
            plt.plot(x, y, marks[index], color=colors[index], label=labels[index])
    
    plt.legend(loc='best', fontsize=18)
    if plot_num == 4:
        plt.tight_layout()
        plt.show()
```

```python
plt.subplots(1, 4, figsize=(30,5))

labels = ['Constant', 'Cosine', 'Diminishing', 'Exponential']
colors = ['blue', 'orange', 'red', 'green']
marks = ['-^', '-v', '->', '-<']
smg_different_schedulers = ['SMG_constantLR', 'SMG_cosineLR', 'SMG_diminishingLR', 'SMG_exponentialLR']

path = '/content/gdrive/MyDrive/SMGExperiments/SMG-Fashion-History/'
plot_loss_nn(path, smg_different_schedulers, labels, colors, marks, 'fashionmnist', 1)

path = '/content/gdrive/MyDrive/SMGExperiments/SMG-Cifar10-History/'
plot_loss_nn(path, smg_different_schedulers, labels, colors, marks, 'cifar10', 2)

path = '/content/gdrive/MyDrive/SMGExperiments/SMG-w8a-History/'
plot_loss_non_convex(path, smg_different_schedulers, labels, colors, marks, 'w8a',3)

path = '/content/gdrive/MyDrive/SMGExperiments/SMG-ijcnn1-History/'
plot_loss_non_convex(path, smg_different_schedulers, labels, colors, marks, 'ijcnn1', 4)
```
