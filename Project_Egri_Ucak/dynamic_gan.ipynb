```python
from hubconf import PGAN
```

```python
import matplotlib.pyplot as plt # For plotting
import numpy as np              # NumPy, for working with arrays/tensors 
```

```python
model = PGAN(True, model_name="celebAHQ-512")
```

```python
print(model)
```

```python
inputRandom, randomLabels = model.buildNoiseData(16)
```

```python
fig = plt.figure(figsize=(32, 32))
out = model.test(inputRandom, getAvG=True, toCPU=True)
for i in range(4):
    for j in range(4):
        fig.add_subplot(4, 4, i * 4 + j + 1)
        plt.imshow(np.transpose(out[i * 3 + j], (1, 2, 0)), interpolation='nearest')
        # Show results on a 2x2 grid
        # plt.imshow(np.transpose(out[0], (1, 2, 0)), interpolation='nearest')
plt.show()
```
